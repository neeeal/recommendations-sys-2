```python
import pandas as pd
```

```python
data= pd.read_csv('datasets\imdb_movies.csv')
```

```python
data.head()
```

```python
summary=data['overview'].values

print(summary)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")

```

```python
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python

for n,name in enumerate(summary):
    temp = name.lower().split(" ")
    temp = [''.join([letter for letter in word if letter.isalnum()]) for word in temp]
    temp = [word for word in temp if word not in stop_words]
    temp = ' '.join(temp)
    summary[n] = temp
```

```python
summary
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
#Building BOW
model = Tokenizer()
model.fit_on_texts(summary)
bow = model.texts_to_matrix(summary, mode='count') # bag of words
bow_keys=f'Key : {list(model.word_index.keys())}'
```

```python
print(bow_keys)
```

```python
## CREATING DATAFRAME FOR TRAINING
df_name=pd.DataFrame(data['names'])
df_name.rename(columns = {'names':'Title'}, inplace = True)
df_count=pd.DataFrame(bow)
frames=[df_name,df_count]
result=pd.concat(frames,axis=1)
result = result.drop_duplicates(subset='Title', keep='last') ## REMOVING DUPLICATES
result=result.set_index('Title')
result=result.drop([0], axis=1)
```

```python
print(result.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#Calculating cosine similarity
df_name=df_name.convert_dtypes(str)
temp_df=df_name['Title']
temp_df = temp_df.drop_duplicates( keep='last').reset_index().drop(columns='index')
print(temp_df.shape)
```

```python
sim_df = pd.DataFrame(cosine_similarity(result, dense_output=True))
```
